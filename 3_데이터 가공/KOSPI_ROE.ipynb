- KOSPI 5개년 ROE 지수 확인

```python
df = pd.read_excel("C:\Workspace\HK_Project\data\KOSPIROE5years.xlsx")
df.head()
```

```python
df.replace(' ', '', inplace=True)
df
```

```python
df.replace('', np.nan, inplace=True)
df
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
df
```

```python
df.info()
```

```python
df[2021] = df[2021].astype(float)
df.info()
```

```python
col = [2017, 2018, 2019, 2020, 2021]
```

```python
# ROE 11이상 
df_ROE_11_upper = df[(df[2017] >= 11) & (df[2018] >= 11) & (df[2019] >= 11) & (df[2020] >= 11) & (df[2021] >= 11)]
df_ROE_11_upper
```

```python
# '평균' 열 추가
df_ROE_11_upper["average"] = df_ROE_11_upper.mean(axis=1)
df_ROE_11_upper
```

```python
sns.barplot(
    data = df_ROE_11_upper,
    x = "종목명",
    y = "average"
)
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['font.size'] = 10
plt.xticks(rotation=90)
plt.show()
```

```python
df_ROE_11_upper.info()
```
