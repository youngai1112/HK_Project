```python
import pandas as pd
```

```python
# 열의 갯수 (컬럼의 갯수는 4개)
df = pd.DataFrame({'foo' : ['one', 'one', 'one', 'two', 'two', 'two'],
                'bar' : ['a', 'b', 'c', 'a', 'b', 'c'],
                'baz' : [1,2,3,4,5,6],
                'zoo' : ['x', 'y', 'z', 'q', 'w', 't']})
df
```

```python
type(df)
```

```python
# bar을 제외한 나머지 특성에 대해서 a, b, c를 달아줌
#   데이터 중복을 막기위해
df_after = df.pivot(index=None, columns='bar', values=None)
df_after
```

```python
df.pivot(index=None, columns='bar', values='foo')
# df.pivot(index=None, columns='bar')['foo']
```

```python
df.pivot(index='foo', columns='bar')
```
