### OHLCV
- open, high, low, close, volume (당일 시가, 고가, 저가, 종가, 거래량)
- pykrx에서 OHLCV 출력에 대해 제공하는 형태
    - 전체 종목 시세 출력
    - 단일 종목의 기간 단위로 출력

1. 전체 종목 시세 출력

```python
from pykrx import stock

df = stock.get_market_ohlcv_by_ticker("20201021")
print(df.head(10))
```

2. 단일 종목의 기간 단위로 출력

```python
df = stock.get_market_ohlcv_by_date("20220101", "20220225", "005930")
print(df.head(10))
```

```python
# 특정군의 모든 데이터들을 출력하거나, 작업할 때 사용
for ticker in stock.get_market_ticker_list():
    df = stock.get_market_ohlcv_by_date("20220101", "20220201", ticker)
    print(df.head())
    # time.sleep(1)
```
