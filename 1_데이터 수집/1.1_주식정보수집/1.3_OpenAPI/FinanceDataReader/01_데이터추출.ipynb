4. 종목 데이터 가져오기

```python
# Apple
aapl = fdr.DataReader('AAPL')   # 전체 기간
aapl_2022 = fdr.DataReader('AAPL', '2022')  # 2022 데이터만
aaple_2000_2022 = fdr.DataReader('AAPL', '2000-01-01', '2022-01-01')    # 2000-01-01 ~ 2022-01-01 데이터만
```

```python
aapl
```

```python
aapl_2022.tail()
```

```python
aaple_2000_2022.head()
```
